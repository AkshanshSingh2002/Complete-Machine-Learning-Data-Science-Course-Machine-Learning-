### Data Imputation

```python
fig, ax = plt.subplots(4, 2, figsize=(15, 10))
sns.histplot(dataframe['Pregnancies'], ax=ax[0, 0], kde=True)
sns.histplot(dataframe['Glucose'], ax=ax[0, 1], kde=True)
sns.histplot(dataframe['BloodPressure'], ax=ax[1, 0], kde=True)
sns.histplot(dataframe['SkinThickness'], ax=ax[1, 1], kde=True)
sns.histplot(dataframe['Insulin'], ax=ax[2, 0], kde=True)
sns.histplot(dataframe['BMI'], ax=ax[2, 1], kde=True)
sns.histplot(dataframe['DiabetesPedigreeFunction'], ax=ax[3, 0], kde=True)
sns.histplot(dataframe['Age'], ax=ax[3, 1], kde=True)
plt.tight_layout()
plt.show()
```

From the above distribution chart we can see that we have to perform impurtation in all of the fields. <br>
*   Pregnancies ---> Median
*   Glucose ---> Mean
*   BloodPressure ---> Mean
*   SkinThickness ---> Median
*   Insulin ---> Median
*   BMI ---> Mean
*   DiabetesPedigreeFunction ---> Median
*   Age ---> Median

```python
# Imputation of 0's
dataframe['Pregnancies'] = dataframe['Pregnancies'].replace(0, dataframe['Pregnancies'].median())
dataframe['Glucose'] = dataframe['Glucose'].replace(0, dataframe['Glucose'].mean())
dataframe['BloodPressure'] = dataframe['BloodPressure'].replace(0, dataframe['BloodPressure'].mean())
dataframe['SkinThickness'] = dataframe['SkinThickness'].replace(0, dataframe['SkinThickness'].median())
dataframe['Insulin'] = dataframe['Insulin'].replace(0, dataframe['Insulin'].median())
dataframe['BMI'] = dataframe['BMI'].replace(0, dataframe['BMI'].mean())
dataframe['DiabetesPedigreeFunction'] = dataframe['DiabetesPedigreeFunction'].replace(0, dataframe['DiabetesPedigreeFunction'].median())
dataframe['Age'] = dataframe['Age'].replace(0, dataframe['Age'].median())
```

### Outlier Detection

```python
## X -> input features,y -> output
X = dataframe.drop('Outcome', axis=1)
y = dataframe['Outcome']
```

```python
## BOX PLOT
fig, ax = plt.subplots(figsize=(15, 15))
sns.boxplot(data=dataframe, ax=ax)
plt.show()
```

#### Approach of quantiles to remove the outliers

```python
cols = X.columns.tolist()
for col in cols:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    mask = (X[col] >= lower_bound) & (X[col] <= upper_bound)
```

```python
X_no_outliers = X[mask]
y_no_outliers = y[mask]
```

```python
y_no_outliers.value_counts()
```

#### Handeling imbalanced data

```python
## Normalizing the data (resuce the biasness of the data)
# Standard Normal Form: Mean = 0 and Standard Deviation = 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_no_outliers)
```
